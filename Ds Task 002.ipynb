```python
from colorama import Fore,Back
import pandas as pd
df = pd.read_html("https://stats.espncricinfo.com/ci/content/records/335433.html")[0]
```

```python
df_column_colors=['red','blue','green','yellow','orange']

def apply_color(cells):
    color = df_column_colors[df.columns.get_loc(cells.name)]
    colors = []
    for cell in cells:
        if cell == '':
            colors.append('')
        else:
            colors.append('background-color: %s' % color)
    return colors

df.style.apply(apply_color)
```
